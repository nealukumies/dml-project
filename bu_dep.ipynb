# Traffic Accident Analysis 

This notebook is for analyzing traffic accident data provided by the Finnish Transport Infrastructure Agency in order to find factors associated with accident severity and to discover typical accident patterns. This information could be used to improve road safety by identifying high-risk conditions and locations. The goal is to use machine learning techniques to model following research questions:

### Research Question 1: Can accident severity be predicted based on the available features?
### Research Question 2: How much do weather conditions affect accident severity?
### Research Question 3: Are there identifiable accident profiles (clusters)?

The first two are classification problems, while the third is an unsupervised learning problem. Suitable models for classification include Decision Trees, Random Forests and Logistic Regression while for clustering the K-Means algorithm can be used. 

The requirement for this analysis is to understand the data, preprocess it and apply machine learning models to answer the research questions. The analysis will be performed using Python and libraries such as pandas, scikit-learn and matplotlib. The data will be a constraining factor as the data model is complex and varies between years making it difficult to combine data from multiple years. It is however vital to combine several years worth of data as the hypothesis is that weather conditions will have  a significant impact on accident severity and this can only be tested with a large dataset that includes a variety of weather conditions. 

# Deployment

These models could be deployed as a supporting tool for decision making in traffic safety planning and management. The Finnish Transport Infrastructure Agency could include these in a dashboard that provides insights on accident patterns and severity predictions based on current conditions. It would then be possible to lower accident rates by for example issuing warnings or lowering speed limits in high-risk conditions. 

As it stands, the models do not have the accuracy required for deployment but with further work and more data, it is possible that they could be improved to a level where they could be used in a real-world application. The data is updated annually and the models would need to be retrained with new data to maintain their accuracy. Also, as the data model has been updated in 2022 it is possible that combining data from more years would make the models more accurate, but it would also require more work in data preprocessing. With the updated model it is however possible, that the later years of data are better filled and more accurate, which could lead to better models.
